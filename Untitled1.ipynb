2. How does broadcasting work in NumPy?
  - Broadcasting applies specific rules to determine whether two arrays can be aligned for operations:
  
  Check Dimensions: Ensure the arrays have the same number of dimensions or expandable dimensions.
  
  Dimension Padding: If arrays have different numbers of dimensions, the smaller array is left-padded with ones.
  
  Shape Compatibility: Two dimensions are compatible if:
  They are equal, or
  One of them is 1.
  
  If these conditions are not met, a ValueError is raised.

  Example:
   
   import numpy as np
   arr = np.array([1, 2, 3])
   res = arr + 1  # Adds 1 to each element
   print(res)

4. Explain the use of the groupby() method in Pandas?

17. Explain the difference between apply() and map() in Pandas
   - apply() and map() are both used for applying functions to Pandas Series, but they operate differently:
 map():
Works element-wise on a Series.
It can accept a function, a dictionary, or a Series.
If a function is provided, it's applied to each element in the Series.
If a dictionary or Series is provided, it's used to map values from the Series to new values.
It is optimized for element-wise operations and can be faster than apply() for simple transformations.
apply():
Can be used on both Series and DataFrames.
When used on a Series, it applies a function to each element, similar to map() when used with a function.
When used on a DataFrame, it can apply a function to either the rows (axis=1) or columns (axis=0).
It is more flexible than map() as it can handle more complex operations, including those that require access to multiple columns or rows.
It can also be used to return Series or DataFrames, allowing for more complex transformations.

21. Why is NumPy’s array slicing faster than Python’s list slicing
   - NumPy array slicing is faster than Python list slicing due to the way data is stored and processed. NumPy arrays store elements of the same data type in contiguous memory blocks. This allows for efficient access and manipulation of data, as the location of each element can be calculated directly. In contrast, Python lists store elements as pointers to memory locations, which can be scattered throughout memory. When slicing a list, Python needs to follow these pointers to collect the elements, which is a slower process.
   
   Additionally, NumPy operations, including slicing, are implemented in C and Fortran, which are compiled languages. This allows for optimized execution and vectorized operations, where the same operation is performed on multiple elements simultaneously. Python lists, on the other hand, rely on Python's interpreter, which has more overhead and slower execution for numerical computations.

29. Write a program to perform element-wise multiplication of two NumPy arrays



```python
import numpy as np

def multiply_arrays(array1, array2):

  if array1.shape != array2.shape:
    print("Error: Arrays must have the same shape for element-wise multiplication.")
    return None
  return np.multiply(array1, array2)

if __name__ == '__main__':
  array1 = np.array([[1, 2, 3], [4, 5, 6]])
  array2 = np.array([[7, 8, 9], [10, 11, 12]])

  result_array = multiply_arrays(array1, array2)

  if result_array is not None:
    print("Array 1:")
    print(array1)
    print("\nArray 2:")
    print(array2)
    print("\nElement-wise product:")
    print(result_array)
```

31. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

```python
import pandas as pd
import numpy as np

data = {'col_A': np.random.randint(0, 100, 20),
        'col_B': np.random.rand(20),
        'col_C': ['label_' + str(i) for i in range(20)]}
df = pd.DataFrame(data)

threshold = 50

filtered_df = df[df['col_A'] > threshold]

print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (col_A > {}):".format(threshold))
print(filtered_df)
```

32. Create a histogram using Seaborn to visualize a distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset (e.g., penguins dataset)
penguins = sns.load_dataset('penguins')

# Create a basic histogram
sns.histplot(data=penguins, x='flipper_length_mm')
plt.show()

# Customize the histogram
sns.histplot(data=penguins, x='flipper_length_mm', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

35. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
data = pd.DataFrame({
    'x': np.random.randn(100),
    'y': np.random.randn(100),
    'z': np.random.randn(100),
    'color': np.random.choice(['A', 'B', 'C'], 100),
    'size': np.random.rand(100) * 20
})

fig = px.scatter_3d(
    data,
    x='x',
    y='y',
    z='z',
    color='color',
    size='size',
    opacity=0.7,
    title='3D Scatter Plot Example'
)

fig.show()
```
